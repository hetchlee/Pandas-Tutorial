```python
import requests
import numpy as np
import scipy as sp
import pandas as pd
import datetime
import json
```

```python
#获取数据,并保存到iris.txt中
r = requests.get("http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
with open('iris.txt','w') as f:
    f.write(r.text)
```

```python
#打开文本文件，读取iris.txt数据

with open('iris.txt','r') as f:
    lines = f.readlines()
    
for line in lines:
    print line
```

```python
#使用pandas读取txt数据，使用读取CSV的读取方法，方便后面快速识别结构化数据
irisdata = pd.read_csv('iris.txt',header = None)
irisdata
```

```python
#上面已经将txt中的数据按照结构化的行列形式展现出来了
#那么如何给每一列取个名字呢
colnames = ['sepal_length','sepal_width','petal_length','petal_width','class']
irisdata.columns = colnames
irisdata
```

```python
#如何对数据进行过滤呢，比如说我们需要找到那些width为最大值的数据
irisdata[irisdata['petal_width']==irisdata['petal_width'].max()]
```

```python
#如何对数据进行切片呢,：：30的意思是从0开始到结束，每隔30个数取一个数
#：2的意思是前两列数据进行显示
irisdata.iloc[::30,:2]
```

```python
#数据集中的列表class有三类，那么如何获取每一类到底有多少个呢？
#以及如何获取一些基本的统计数据，如最大最小值，中位数，平均数等
#下面的{0:>5}表示间距
print irisdata['class'].value_counts()
for x in xrange(4):
    s = irisdata.iloc[:,x] #取每一列
    print '{0:<12}'.format(s.name.upper()),"Statistics",'{0:>5} {1:>5} {2:>5} {3:>5}'.format(s.max(),s.min(),round(s.mean(),2),round(s.std(),2))
    
```

```python
#如何做快速mapreduce,如何分组
slogs = lambda x:sp.log(x)*x
entpy = lambda x:sp.exp((slogs(x.sum())-x.map(slogs).sum())/x.sum())
irisdata.groupby('class').agg(entpy)
```
